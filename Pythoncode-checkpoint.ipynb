### Missing Data:

```python
table.dropna(axis=1, how='all')
```

```python
hun = table[table.country == 'Hungary']
aut = table[table.country == 'Austria']
```

```python
hun
```

```python
aut
```

```python
plt.plot(hun.year, hun.cn)
plt.plot(aut.year, aut.cn)
plt.xlabel('Years')
plt.ylabel('Capital Stock (in current PPPs, mil. 2017 $US)')
plt.title('The development of Austrian and Hungarian capital stock, 1950-2019')
plt.legend(['Hungary', 'Austria'])
plt.show()
```
